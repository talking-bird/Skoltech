## Task 1. Bifurcation map
$x_{n+1}\leftarrow rx_n(1-x_n)$\
$x_0\leftarrow \text{rand}(0,1)$\
$r\leftarrow\text{const}$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def bif_map(r,x):
  return r*x*(1-x)

x_1=np.linspace(0,1,100)
```

### 1.Implement the map, plot the evolution of x

```python
plt.figure(figsize=(6,4))
plt.plot(x_1, bif_map(1,x_1),'k')
plt.title("Map for $r=1$",fontsize=15)
plt.xlabel("$x$",fontsize=15)
plt.ylabel("$rx(1-x)$",fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

```python
def plot_recursion(r, x0, n):
  t=np.linspace(0,1)
  #plot of the map
  plt.plot(t,bif_map(r,t),'k',lw=2)
  #line x=y
  plt.plot([0,1],[0,1],'k',lw=2)

  x=x0
  for i in range(n):
    y = bif_map(r,x)
    plt.plot([x,x],[x,y],'k', lw=1)
    plt.plot([x,y],[y,y],'k', lw=1)
    plt.plot([x], [y], 'ok', ms=10,alpha=(i + 1) / n)
    x = y

  plt.xlim(0,1)
  plt.ylim(0,1)
  plt.title(f'$X$ evolution : $r={r}, \, x_0={x0}$',fontsize=15)
  plt.xticks(fontsize=15)
  plt.yticks(fontsize=15)
```

```python
plt.figure(figsize=(6,6))
plot_recursion(2.5, 0.1, 10)
```

```python
plt.figure(figsize=(6,6))
plot_recursion(3.5, 0.1, 10)
```

### 2.Create a linspace of rs, for every r save the last m("last" in my case) values of x after the first n("first" in my case) values (can be m=200, x=200)

```python
n = 10000
r = np.linspace(0,4.0, n)
first = 900
last = 100
#initial condition
x0 = np.random.random()
x = x0 * np.ones(n)
```

### 3.Plot the bifurcation map

```python
plt.figure(figsize=(14,11))
for i in range(first+last):
  x = bif_map(r, x)
  if i>= first:
    plt.scatter(r, x,c='black', alpha=.05,s=0.1)
plt.xlim(0,4)
plt.xlabel("$r$",fontsize=30)
plt.ylabel("$x_\infty$",fontsize=30)
plt.title('Bifurcation map, $x \in [0,4]$',fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.show()
```
